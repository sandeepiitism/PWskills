### linspace
points in same samling difference, linear relationship

```python
x=np.linspace(0,10,100)
x
```

```python
plt.plot(x)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x), '--')
```

```python
plt.figure()  ## plt is the canva
plt.subplot(1,2,1)   ## create 1 page where 2 subpart and access the first one --> 1x2 canva
plt.plot(x, np.sin(x))

plt.subplot(1,2,2)         ### access the 2 one
plt.plot(x, np.cos(x))

plt.show()  ## show me that page
```

```python
plt.figure()  ## plt is the canva
plt.subplot(2,2,1)   ## create 2 page where 2 subpart and access the first one --> 2x2 canva
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)       
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)       
plt.plot(x, np.tan(x))

plt.subplot(2,2,4)       
plt.plot(x, np.cos(x+1))
```

```python
plt.figure()  ## plt is the canva
plt.subplot(2,1,1)   ## create 2 page where 2 subpart and access the first one --> 2x1 canva
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)       
plt.plot(x, np.cos(x))
```

```python
## color of the plot

plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='yellow')
plt.plot(x, np.sin(x-2), color=(1.0, 0.5, 0))
```

```python
plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1,15)

plt.show()
```

```python
plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1,15)
plt.axis('equal')       ### to fit the plot 
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1,15)
plt.grid()   ### adding grid
plt.axis('equal')       ### to fit the plot 
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1,15)
plt.title("Test Plot")  ### add title
plt.axis('equal')       ### to fit the plot 
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color='blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1,15)
plt.title("Test Plot")  ### add title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.axis('equal')       ### to fit the plot 
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color='blue', label='sinx')
plt.plot(x, np.cos(x-0), color='red', label='cosx')
plt.ylim(-1.5, 1.5)
plt.xlim(1,15)
plt.title("Test Plot")  ### add title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.axis('equal')       ### to fit the plot 
plt.legend()         ### legend used to call plt.plotlabels
plt.show()
```

### Histogram

for contineous data, to know the frequency use histogram

```python
data = np.random.randn(1000)
plt.hist(data)
plt.show()

### by default there will be 10 bins
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, histtype='barstacked')
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, histtype='bar', color='green')
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, histtype='step', color='red')
plt.show()
```

### Bar plot

categorical plot

```python
x = ['a', 'b', 'c', 'd']
y=[3,5,4,7]
```

```python
plt.figure(figsize=(6,3))
plt.bar(x,y,color='red')
```

```python
plt.figure(figsize=(6,3))
plt.barh(x,y,color='green')
```

```python
x=np.random.rand(20)
y=np.random.rand(20)
z=np.random.rand(20)

fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c="red", alpha=0.9)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
